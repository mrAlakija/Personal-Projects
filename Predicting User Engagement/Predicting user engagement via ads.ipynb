# Predicting User Engagement Via Ads

In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

# Import Data analysis and visualization Libraries



```python
# import numpy
import numpy as np

# import pandas
import pandas as pd

#import matplotlib
import matplotlib.pyplot as plt

# import seaborn and set its style to 'darkgrid'
import seaborn as sns
sns.set(style='darkgrid')
```

# Get the Data


```python
# read the csv file
ad_data = pd.read_csv('datasets/advertising.csv')

# Check the first 5 records
ad_data.head()
```

# Summary Statistics

```python
print("Summary of the Data: \n")
print(ad_data.info(), "\n")

print("General descriptive Statistics: \n")
ad_data.describe()
```

# Exploratory Data Analysis

Using seaborn, lets carry out some Exploratory Data Analysis to better understand the data. 

   ## distribution plot showing the univariate distribution of user Age

```python
sns.distplot(ad_data.Age, kde= False, bins=30, color='r');
```

## jointplot showing Area Income versus Age

```python
sns.jointplot(data = ad_data, x = ad_data['Age'], y = ad_data['Area Income']);
print(' From the figure below, Users aged 25 - 40 make the most income \n The older you get the more a user''s income increases')
```

## jointplot showing the kde distributions of Daily Time spent on site vs. Age

```python
sns.jointplot(data = ad_data, x = 'Age', y = 'Daily Time Spent on Site', kind = 'kde', color="r",  height=7);
print('Figue shows that Users aged 30 - 40 spend the most time on a site')
```

## jointplot of 'Daily Time Spent on Site' vs. 'Daily Internet Usage'

```python
sns.jointplot(data = ad_data, x = 'Daily Time Spent on Site', y = 'Daily Internet Usage', color = 'g');
print('There are two visible clusters. one at the begining of the diagram and another at the tail end')
```

## pairplot with the hue defined by the 'Clicked on Ad' column feature

```python
sns.pairplot(data = ad_data, hue='Clicked on Ad');
```

# Logistic Regression

As the target variable is **binary**, this is a classification problem hence we would use a Logistic regression algorithm to solve this problem.

## selecting the features and target variables

```python
# selecting only the numerical features 
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
# selecting only the target variable
y = ad_data['Clicked on Ad']
```

## splitting the data into test and train sets

```python
#import train_test_split
from sklearn.model_selection import train_test_split

# split the data into 70/30 ratio`
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)

# import the regression model
from sklearn.linear_model import LogisticRegression

# instantiate a regression model
logreg = LogisticRegression(solver='lbfgs')

# fitting the model on the training set
logreg.fit(X_train, y_train)
```

## Predictions and Evaluations


```python
# predicting 
y_pred = logreg.predict(X_test)

# evaluating our models performance
from sklearn.metrics import classification_report

# printing the classification report
print(classification_report(y_test,  y_pred))
```
